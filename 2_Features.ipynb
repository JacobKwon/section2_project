```python
data = pd.read_csv('./17rc_eda.csv')
data.head()
```

```python
#중복값 확인
data.duplicated().sum()
```

```python
Use_feature = ['NF_CA','NF_CAROT','NF_NA','NF_EN','NF_K','NF_FE','NF_NIAC','NF_PHOS','NF_WATER','NF_RETIN','NF_VITC','NF_VA','NF_PROT','NF_CHO','NF_FAT']

#INQ 구하기
# 1000칼로리 대 비율.
data['INQ'] = 0

#INQ = 1000kal당 영양소 섭취량
#칼로리 = 단백질 * 4 + 탄수화물 * 4 + 지방 * 9 / 1000

#INQ 계수 구하기
# data['INQ'] = (data['NF_PROT']*4) + (data['NF_CHO']*4) + data['NF_FAT']*9
#칼로리 데이터가 있었음
data['INQ'] = data['NF_EN'] / 1000
```

#INQ 구하기
#각 영양소 섭취량 / 각 권장량 * INQ 계수 -> 영양소 mean값

#각 영양소 권장량
NF_CA = 600mg
NF_CAROT = 6000IU
NF_NA = 2000mg
NF_EN = 2400kal
NF_K = 3500mg
NF_FE = 10mg
NF_NIAC = 14mg
NF_PHOS = 700mg
NF_WATER = 2500ml
NF_RETIN = 900ug / 3000 IU
NF_VITC = 100mg
NF_VA = 2500IU
NF_PROT = 70g
NF_CHO = 2000kal
NF_FAT = 13g


```python
data.describe()
```

```python
#INQ값 구하기
data['NF_CA'] = (data['NF_CA']*data['INQ']) / 600
data['NF_CAROT'] = (data['NF_CAROT']*data['INQ']) / 6000
data['NF_NA'] = (data['NF_NA']*data['INQ']) / 2000
data['NF_EN'] = (data['NF_EN']*data['INQ']) / 2400
data['NF_K'] = (data['NF_K']*data['INQ']) / 3500
data['NF_FE'] = (data['NF_FE']*data['INQ']) / 10
data['NF_NIAC'] = (data['NF_NIAC']*data['INQ']) / 14
data['NF_PHOS'] = (data['NF_PHOS']*data['INQ']) / 700
data['NF_WATER'] = (data['NF_WATER']*data['INQ']) / 2500
data['NF_RETIN'] = (data['NF_RETIN']*data['INQ']) / 900
data['NF_VITC'] = (data['NF_VITC']*data['INQ']) / 10
data['NF_VA'] = (data['NF_VA']*data['INQ']) / 900
data['NF_PROT'] = (data['NF_PROT']*data['INQ']) / 70
data['NF_CHO'] = (data['NF_CHO']*data['INQ']) / 2000
data['NF_FAT'] = (data['NF_FAT']*data['INQ']) / 13
```

```python
data.head()
```

```python
data['INQ'] = data[Use_feature].sum(axis=1) / len(Use_feature)
```

```python
data.head()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
data.describe()
```

```python
data.to_csv('./18rc_eda.csv', index=False)
```
