```python
#3 / test-4
def eda_process(tmp):
    #1. age_month 결측치 처리
    tmp['age_month'].fillna(tmp['age']*12, inplace=True)
    #df['age_month'].isnull().sum()
    
    #2. 사용할 피쳐 남기고 제거
    features = ['ID','ID_fam','year','region','psu','apt_t','sex',
                'age_month','incm5','ho_incm5','edu','occp','kstrata',
                'N_BFD_Y','N_DAY','N_DCODE','N_DNAME','N_DIET','N_DIET_WHY',
                'N_DUSUAL','N_FCODE','N_MTYPE','N_MEAL','N_MEAL_P','N_MEAL_T',
                'N_FNAME','N_KINDG1','N_FM_WT','N_PRG','N_WAT_C','NF_CA',
                'NF_FAT','NF_CAROT','NF_CHO','NF_NA','NF_EN','NF_K','NF_FE',
                'NF_INTK','NF_NIAC','NF_PHOS','NF_PROT','NF_WATER','NF_RETIN',
                'NF_VITC','NF_VA','NF_tdf']
    tmp = tmp[tmp.columns.intersection(features)]
    
    #3. N_DT_DS, N_DT_ETC 결측치 채우기 (바이너리 0으로) / 분류가 아니라 설문이라 drop
#     tmp['N_DT_DS'].fillna(0, inplace=True)
#     tmp['N_DT_ETC'].fillna(0, inplace=True)
    
    #4. 특정행 결측치 제거
    tmp.dropna(
        subset=[
            'incm5','ho_incm5','edu','occp',
            'N_FCODE','N_FNAME','N_FM_WT','N_KINDG1','NF_INTK','NF_EN','NF_WATER',
            'NF_PROT','NF_FAT','NF_CHO','NF_tdf','NF_CA','NF_PHOS','NF_FE',
            'NF_NA','NF_K','NF_VA','NF_CAROT','NF_RETIN','NF_NIAC','NF_VITC'
        ],
        inplace=True)
    
#     #5. euc-kr로 디코드
#     #바이트 -> euc-kr로 디코딩 할 피쳐
#     decode_utf8_feature = ['ID','ID_fam','psu','N_DUSUAL','N_MEAL_T',
#                       'N_DCODE','N_FCODE','N_KINDG1']
#     decode_euckr_feature = ['N_DAY','N_DNAME','N_FNAME']
    
#     for x in decode_utf8_feature:
#         print(x)
#         df[x] = df[x].str.decode('utf-8')#cp949
# #         df[x] = df[x].str.encode('utf-8')

#     for x in decode_euckr_feature:
#         df[x] = df[x].str.decode('euc-kr')

#     df[decode_euckr_feature] = map(lambda x: x.encode('ascii','strict'), df[decode_euckr_feature])
#     df[decode_euckr_feature] = map(lambda x: x.str.decode('utf-8'), df[decode_euckr_feature])
    

    #약 4시간 정도 공부한 듯..
    #파일마다 특정 인코딩 형태로 되어있을 확률이 높으니,
    #csv로 저장한 후 utf-8 형태로 불러오는 것이 최선일 것 같다.
    
    #중복제거
    tmp.drop_duplicates(keep='first', inplace=True)
    
    print('DataFrame Shape : ',tmp.shape)
    
    return tmp
```

```python
#4
def readSas_CallEDA(filenames):
    
    for x in filenames:
        fname = x.split('.')
        tmp_df = pd.read_sas(f'./{x}', format='sas7bdat')
        
        print('file name : ',x,
              ' and columns count : ', len(tmp_df.columns.tolist()))
#         print(fname[0])
    
        if(fname[0] == 'hn19_24rc'):
            print('19년도 NF_VA_RAE -> NF_VA')
            tmp_df.rename(columns={'NF_VA_RAE':'NF_VA'}, inplace=True)
        
        if((fname[0] == 'hn11_24rc') | (fname[0] == 'hn12_24rc')):
            print('10~12 NF_FIBER -> NF_tdf 변경')
            tmp_df.rename(columns={'NF_FIBER':'NF_tdf'}, inplace=True)
#             print(tmp_df.columns.tolist())
        
        #예외가 생김ㅠㅠ
        if((fname[0] == 'hn15_24rc') |
           (fname[0] == 'hn16_24rc') |
           (fname[0] == 'hn17_24rc') |
           (fname[0] == 'hn18_24rc') |
           (fname[0] == 'hn19_24rc') |
           (fname[0] == 'hn20_24rc')):
            print('10~12 NF_FIBER -> NF_tdf 변경')
            tmp_df.rename(columns={'NF_TDF':'NF_tdf'}, inplace=True)
        
        # EDA FUNCTION CALL
        tmp_df = eda_process(tmp_df)
        
        tmp_df.to_csv(f'./{fname[0]}.csv',index=False)
        
        #메모리때문에 루프 한번 돌면 지워준다.
        del tmp_df

        
```
